```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.patches as mpatches

data = pd.read_csv('VD_Participant_1_22_Data.csv', sep=';')

correct_data = data[data['outcomeArrayEXP2'] == 1]['elapsed_timeEXP2']
wrong_data = data[data['outcomeArrayEXP2'] == 0]['elapsed_timeEXP2']

mean_correct = correct_data.mean()
mean_wrong = wrong_data.mean()

sem_correct = correct_data.sem()
sem_wrong = wrong_data.sem()

t_stat, p_val = ttest_ind(correct_data, wrong_data)

fig, ax = plt.subplots(figsize=(8, 6))
labels = ['Correct Decisions', 'Wrong Decisions']
x = np.arange(len(labels))
width = 0.35

color1 = '#1874CD'
color2 = '#CD2626'

rects1 = ax.bar(x, [mean_correct, mean_wrong], width, yerr=[sem_correct, sem_wrong], capsize=5, color=[color1, color2])

ax.set_ylabel('Mean Reaction Time')
ax.set_title('Mean Reaction Time for Correct and Wrong Decisions Across All Participants')
ax.set_xticks(x)
ax.set_xticklabels(labels)

legend_handles = [
    mpatches.Patch(color=color1, label=f'Mean (Correct): {mean_correct:.2f}  {sem_correct:.2f}'),
    mpatches.Patch(color=color2, label=f'Mean (Wrong): {mean_wrong:.2f}  {sem_wrong:.2f}'),
    mpatches.Patch(color='white', label=f'p-value: {p_val:.2f}')
]

legend = plt.legend(handles=legend_handles, loc='upper left', fontsize=12, frameon=True)
legend.get_frame().set_facecolor('white')

plt.show()
```
